```python
def train_val_datasets(TRAINING_DIR, VALIDATION_DIR,TEST_DIR, TEST_HR, TEST_SR_DEFAULT, TEST_SR_PIXELDOMINANT, TEST_SR_MIDBOOSTED, TEST_SR_DEFAULTV2, TEST_SR_DEFAULTV3, TEST_SR_DEFAULTV4, TEST_SR_DEFAULTV6, TEST_SR_DEFAULTV17, TEST_SR_DEFAULTV9, TEST_SR_DEFAULT_X2, TEST_SR_PIXELDOMINANT_X2, TEST_LR):
    train_dataset = tf.keras.utils.image_dataset_from_directory( 
        directory=TRAINING_DIR,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",
    ) 
    
    validation_dataset = tf.keras.utils.image_dataset_from_directory( 
        directory=VALIDATION_DIR,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",
    ) 

    test_dataset = tf.keras.utils.image_dataset_from_directory( 
        directory=TEST_DIR,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",
    ) 

    test_hr = tf.keras.utils.image_dataset_from_directory( 
        directory=TEST_HR,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",
    )
    test_sr_default = tf.keras.utils.image_dataset_from_directory( 
        directory=TEST_SR_DEFAULT,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",
    )
    test_sr_pixeldominant = tf.keras.utils.image_dataset_from_directory( 
        directory=TEST_SR_PIXELDOMINANT,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",
    )
    test_sr_midboosted = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_MIDBOOSTED,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_sr_defaultv2 = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_DEFAULTV2,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_sr_defaultv3 = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_DEFAULTV3,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_sr_defaultv4 = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_DEFAULTV4,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_sr_defaultv6 = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_DEFAULTV6,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_sr_defaultv17 = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_DEFAULTV17,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_sr_defaultv9 = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_DEFAULTV9,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_sr_default_x2 = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_DEFAULT_X2,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_sr_pixeldominant_x2 = tf.keras.utils.image_dataset_from_directory( 
        directory=  TEST_SR_PIXELDOMINANT_X2,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",   
    )
    test_lr = tf.keras.utils.image_dataset_from_directory( 
        directory=TEST_LR,
        batch_size=32,
        image_size=(300,300),
		label_mode='categorical',
        color_mode = "rgb",
        
    )
    return train_dataset, validation_dataset, test_dataset, test_hr, test_sr_default, test_sr_pixeldominant, test_sr_midboosted, test_sr_defaultv2, test_sr_defaultv3, test_sr_defaultv4, test_sr_defaultv6, test_sr_defaultv17, test_sr_defaultv9, test_sr_default_x2, test_sr_pixeldominant_x2,  test_lr
```

**Create Augmentation Layer**

```python
augmentation_layer = tf.keras.Sequential([ 
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.2, fill_mode = 'nearest'),
    tf.keras.layers.RandomTranslation(0.2, 0.2, fill_mode = 'nearest'),
    tf.keras.layers.RandomZoom(0.2, fill_mode = 'nearest')
]) 
```

```python
# Load your model for augmentation
data_augmentor = augmentation_layer

# Take a sample image
sample_image = tf.keras.utils.array_to_img(example_batch_images[0])

images = [sample_image]

# Apply random augmentation 3 times
for _ in range(3):
	image_aug = data_augmentor(tf.expand_dims(sample_image, axis=0))
	image_aug = tf.keras.utils.array_to_img(tf.squeeze(image_aug))
	images.append(image_aug)


fig, axs = plt.subplots(1, 4, figsize=(14, 7))
for ax, image, title in zip(axs, images, ['Original Image', 'Augmented 1', 'Augmented 2', 'Augmented 3']):
    ax.imshow(image, cmap='gray')
    ax.set_title(title)
    ax.axis('off')

plt.show()
```

```python
def augment (image,label):
    image = data_augmentor(image)
    return image, label
```

**Use the tf.keras.applications.efficientnet.preprocess_input function to preprocess images correctly. This function ensures the images are scaled and normalized as required by EfficientNet.**

```python
preprocess_input = tf.keras.applications.efficientnet.preprocess_input
def preprocess(dataset, is_training=False):
    if is_training:
        dataset = dataset.map(augment, num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.map(lambda x, y: (preprocess_input(x), y), num_parallel_calls=tf.data.AUTOTUNE)
    return dataset.cache().shuffle(1000 if is_training else 1).prefetch(tf.data.AUTOTUNE)
```

```python
train_dataset = preprocess(train_dataset, is_training=True)
val_dataset = preprocess(validation_dataset, is_training=False)
test_dataset = preprocess(test_dataset, is_training=False)
test_hr = preprocess(test_hr, is_training=False)
test_sr_default = preprocess(test_sr_default, is_training=False)
test_sr_pixeldominant = preprocess(test_sr_pixeldominant, is_training=False)
test_sr_midboosted = preprocess(test_sr_midboosted, is_training=False)
test_sr_defaultv2 = preprocess(test_sr_defaultv2, is_training=False)
test_sr_defaultv3 = preprocess(test_sr_defaultv3, is_training=False)
test_sr_defaultv4 = preprocess(test_sr_defaultv4, is_training=False)
test_sr_defaultv6 = preprocess(test_sr_defaultv6, is_training=False)
test_sr_defaultv17 = preprocess(test_sr_defaultv17, is_training=False)
test_sr_defaultv9 = preprocess(test_sr_defaultv9, is_training=False)
test_sr_default_x2 = preprocess(test_sr_default_x2, is_training=False)
test_sr_pixeldominant_x2 = preprocess(test_sr_pixeldominant_x2, is_training=False)
test_lr = preprocess(test_lr, is_training=False)
```

**create pre trained model**

```python
import tensorflow as tf
from tensorflow.keras.applications.efficientnet_v2 import EfficientNetV2S
from tensorflow.keras import layers, models

# Load pretrained EfficientNetV2-S tanpa top layer (classifier)
base_model = EfficientNetV2S(
    include_top=False,
    weights='imagenet',
    input_shape=(300, 300, 3),
    pooling='avg'  # GlobalAveragePooling2D
)

# Freeze base model
base_model.trainable = False

model = tf.keras.Sequential([
    tf.keras.Input(shape=(300,300,3)),
    base_model,
    layers.Dense(256, activation='relu'),
    #layers.Dropout(0.3),
    layers.Dense(128, activation='relu'),
    #layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dense(3, activation='linear')                  ,
])

model.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate = 0.00001),
                  loss = tf.keras.losses.CategoricalCrossentropy(from_logits=True),
                  metrics = ['accuracy']
             )
```

```python
import tensorflow as tf
import os

# Ambil semua file path
file_paths = tf.data.Dataset.list_files(TEST_SR_DEFAULTV9 + '/*/*')

def load_img_label_path(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, (300, 300))
    image = image / 255.0

    parts = tf.strings.split(path, os.path.sep)
    label = parts[-2]  # folder sebagai label
    class_names = tf.constant(os.listdir(TEST_SR_DEFAULTV9))
    label_id = tf.where(class_names == label)[0][0]
    label_onehot = tf.one_hot(label_id, depth=len(class_names))

    return image, label_onehot, path

test_sr_defaultv9 = file_paths.map(load_img_label_path).batch(32)

misclassified_images = []
misclassified_true = []
misclassified_pred = []
misclassified_files = []

y_true = []
y_pred = []

for images, labels, file_paths in test_sr_defaultv9:
    predictions = model.predict(images)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(labels.numpy(), axis=1)

    y_true.extend(true_classes)
    y_pred.extend(predicted_classes)

    for idx in range(len(images)):
        if predicted_classes[idx] != true_classes[idx]:
            misclassified_images.append(images[idx])
            misclassified_true.append(true_classes[idx])
            misclassified_pred.append(predicted_classes[idx])
            # Tambahkan nama file
            misclassified_files.append(file_paths[idx].numpy().decode('utf-8'))

# Print file names yang salah klasifikasi
print("File yang salah klasifikasi:")
for i, fname in enumerate(misclassified_files):
    print(f"{i+1}. {fname} | True: {misclassified_true[i]} | Pred: {misclassified_pred[i]}")

# Plot gambar contoh
plt.figure(figsize=(12, 8))
for i in range(min(3, len(misclassified_images))):
    plt.subplot(2, 5, i + 1)
    img = misclassified_images[i].numpy()
    img = (img + 1) / 2.0 if img.min() < 0 else img
    if img.shape[0] == 3:
        img = np.transpose(img, (1, 2, 0))
    plt.imshow(img)
    plt.title(f"{misclassified_files[i].split('/')[-1]}\nT:{misclassified_true[i]} P:{misclassified_pred[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

```
